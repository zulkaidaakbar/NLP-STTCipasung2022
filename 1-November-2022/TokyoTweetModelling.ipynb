Load data Tweeter tentang olimpiade tokyo 2021. Disimpan di variabel df. Filter untuk 6000 tweet saja

```python
#load data 
import pandas as pd 
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('./drive/MyDrive/Colab Notebooks/TopicModelling/TokyoTweetFiltered.csv')
 
# select only 6000 tweets 
df = df[0:6000]
```

Penggunaan bertopik sangat mudah. Hati-hati dengan teks yang dimasukan. Untuk tweeter ini teks nya adalah isi tweet yang berada di kolom "text". Oleh karena itu command nya adalah df.text dan seterusnya

```python
# create model 
 
model = BERTopic(verbose=True)
 
#convert to list 
docs = df.text.to_list()
 
topics, probabilities = model.fit_transform(docs)
```

Bertopic defaultnya kita tidak memasukan jumlah topik. Bertopic akan membuat banyak sekali jumlah topiknya

Kita intip 11 topik teratas

```python
model.get_topic_freq().head(11)
```
